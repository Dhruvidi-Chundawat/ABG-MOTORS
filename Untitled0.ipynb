```python
import pandas as pd
import numpy as np
```

```python
jpn = pd.read_excel("JPN Data.xlsx")
ind = pd.read_excel("IN_Data.xlsx")
```

```python
jpn.head()
```

```python
ind.head()
```

```python

ind.drop(columns=['ID'], inplace=True)
```

```python
jpn.drop(columns=['ID'], inplace=True)
jpn.head()
```

```python
jpn['GENDER'].value_counts(dropna=False)
```

```python
jpn['GENDER'] = jpn['GENDER'].replace({'M': 1, 'F': 0})
ind['GENDER'] = ind['GENDER'].replace({'M': 1, 'F': 0})
```

```python
jpn['GENDER'].unique()
```

```python
X = jpn.drop(columns=['PURCHASE'])
y = jpn['PURCHASE']
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_prob = model.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score

roc_auc = roc_auc_score(y_test, y_prob[:,1])
roc_auc
```

```python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_[0]
})

coefficients
```

```python
import numpy as np

coefficients['Odds_Ratio'] = np.exp(coefficients['Coefficient'])
coefficients
```

```python
# Convert DT_MAINT to datetime
ind['DT_MAINT'] = pd.to_datetime(ind['DT_MAINT'])

# Assume analysis year is 2023 (you can mention this in report)
reference_date = pd.to_datetime("2023-12-31")

# Create AGE_CAR in days
ind['AGE_CAR'] = (reference_date - ind['DT_MAINT']).dt.days
```

```python
ind.drop(columns=['DT_MAINT'], inplace=True)
```

```python
print(X.columns)
print(ind.columns)
```

```python
ind_features = ind[X.columns]
```

```python
ind_scaled = scaler.transform(ind_features)
ind_predictions = model.predict(ind_scaled)
```

```python
ind_scaled = scaler.transform(ind_features)
```

```python
ind_predictions = model.predict(ind_scaled)
```

```python
ind_probabilities = model.predict_proba(ind_scaled)
```

```python
ind['Predicted_Purchase'] = ind_predictions
ind['Purchase_Probability'] = ind_probabilities[:, 1]
```

```python
potential_customers = ind['Predicted_Purchase'].sum()
potential_customers
```

```python
if potential_customers >= 12000:
    decision = "GO: ABG Motors should enter the Indian market"
else:
    decision = "NO-GO: Market entry is not recommended"

decision
```
